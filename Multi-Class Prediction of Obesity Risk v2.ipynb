```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
gender = pd.read_csv('sample_submission.csv')
train_copy = train.copy()
test_copy = test.copy()
train_copy
```

```python
keys_dict= {'Obesity_Type_III': 0,
            'Obesity_Type_II': 1,
            'Normal_Weight' : 2,
            'Obesity_Type_I': 3, 
            'Insufficient_Weight': 4, 
            'Overweight_Level_II': 5, 
            'Overweight_Level_I': 6}
```

```python
def rename(df):
    df.rename(columns={'family_history_with_overweight': 'Family_story',
                     'FAVC': 'High_cal_food', 
                     'FCVC' : 'Veget_food',
                     'NCP' : 'Num_main_meal',
                     'CAEC' : 'Betweem_meal',
                     'SMOKE' : 'Smoke',
                     'CH2O': 'Water_cons',
                     'SCC': 'Calories_monitor',
                     'FAF': 'Phys_activ_frequency',
                     'TUE' : 'Use_devices_time',
                     'CALC': 'Alcohol_cons',
                     'MTRANS': "Transport",
                     'NObeyesdad': 'Target'}, inplace=True)
```

```python
def preprocessing (df):
    # удаляем id
    df = df.drop('id', axis=1)
    # заменяем на численные значения
    df['Gender'] = df['Gender'].replace('Male', 1)
    df['Gender'] = df['Gender'].replace('Female', 0)
    df['Family_story'] = df['Family_story'].replace('no', 0)
    df['Family_story'] = df['Family_story'].replace('yes', 1)
    df['High_cal_food'] = df['High_cal_food'].replace('no', 0)
    df['High_cal_food'] = df['High_cal_food'].replace('yes', 1)
    df['Betweem_meal'] = df['Betweem_meal'].replace('no', 0)
    df['Betweem_meal'] = df['Betweem_meal'].replace('Sometimes', 1/3)
    df['Betweem_meal'] = df['Betweem_meal'].replace('Frequently', 2/3)
    df['Betweem_meal'] = df['Betweem_meal'].replace('Always', 1)
    df['Smoke'] = df['Smoke'].replace('no', 0)
    df['Smoke'] = df['Smoke'].replace('yes', 1)
    df['Calories_monitor'] = df['Calories_monitor'].replace('no', 0)
    df['Calories_monitor'] = df['Calories_monitor'].replace('yes', 1)
    df['Alcohol_cons'] = df['Alcohol_cons'].replace('no', 0)
    df['Alcohol_cons'] = df['Alcohol_cons'].replace('Sometimes', 1/3)
    df['Alcohol_cons'] = df['Alcohol_cons'].replace('Frequently', 2/3)
    df['Alcohol_cons'] = df['Alcohol_cons'].replace('Always', 1)

    #разбиваем столбец "Transport" на 4 столбца
    encoder = OneHotEncoder(sparse_output=False)
    encoded_transport = encoder.fit_transform(df[['Transport']]).astype(int)          # делаем матрицу
    transport_columns = [f"Transport_{i}" for i in range(encoded_transport.shape[1])] # создаем таблицу
    encoded_transport_df = pd.DataFrame(encoded_transport, columns=transport_columns)
    df = pd.concat([df, encoded_transport_df], axis=1)                                # соединяем с нашей
    df = df.drop(['Transport'], axis=1)                                               # удаляем старый столбец

    # Создаем новые столбцы, на основе старых
    df['BMI'] = df['Weight'] / (df['Height'] ** 2)                        # индекс массы тела
    df['Food_health'] = df['Veget_food']/df['High_cal_food']              # соотношение овощей/высоко калорийной пищи
    df['Meal'] = df['Num_main_meal'] + df['Betweem_meal']                 # количество приемов пищи
    df['Activity'] = df['Phys_activ_frequency'] * df['Use_devices_time']  # время физической активности
    return df
```

```python
train_copy = preprocessing(train_copy)
train_copy
```

```python
def columns_with_nan(df):
    b = df.columns.to_numpy()
    c = []
    for col in b:
        if df[col].isnull().values.any():
            c.append(col)
    return c
```

```python
def check(df):
    # Проверка на наличие бесконечных значений
    if np.any(np.isinf(df)):
        print("В данных есть бесконечные значения.")

    # Проверка на наличие слишком больших чисел
    if np.any(np.abs(df) > 1e15):
        print("В данных есть слишком большие числа.")
```

```python
check(test_copy)
```

```python
check(train_copy)
```

```python
def inf_check(df):    # Проверка на бесконечные значения
    inf_values = np.isinf(df)
    if np.any(inf_values):
        indices = np.argwhere(inf_values)
        for idx in indices:
            row_idx, col_idx = idx
            column_name = df.columns[col_idx]
            print(f"Индекс: ({row_idx}, {col_idx}), Название столбца: {column_name}")
```

```python
inf_check(test_copy)
```

```python
def large_check(df):    # Проверка на бесконечные значения
    large_values = np.abs(df) > 1e15
    if np.any(large_values):
        indices = np.argwhere(large_values)
        for idx in indices:
            row_idx, col_idx = idx
            column_name = df.columns[col_idx]
            print(f"Индекс: ({row_idx}, {col_idx}), Название столбца: {column_name}")
```

```python
large_check(test_copy)
```
